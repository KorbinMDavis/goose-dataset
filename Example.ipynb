This is an example to show how to use the GooseDataset class APIs to load data.

```python
import numpy as np
from skimage import io
```

When working in the *goose-dataset* repo, we can simply do the following:

```python
import utils
gd = utils.GooseDataset()
```

```python
(X_train, y_train), (X_test, y_test) = gd.load_data(test_ratio=0.2, grayscale=False, cropped=False, save=False)
```

```python
img = X_train[0,:,:,:]
io.imshow(img)
```

If working outside the goose-dataset repo, we can provide a path to this file and have Python load it:

```python
import importlib.util
spec = importlib.util.spec_from_file_location("utils", "C:\\Users\\Steggie\\git\\goose-dataset\\utils.py")
module = importlib.util.module_from_spec(spec)
spec.loader.exec_module(module)
gd = module.GooseDataset()
```

```python
(X_train, y_train), (X_test, y_test) = gd.load_data(test_ratio=0.2, grayscale=True, cropped=True, save=False)
```

```python
img = X_train[0,:,:,:]
img = img.reshape([img.shape[0], img.shape[1]])
io.imshow(img, cmap="gray")
```
